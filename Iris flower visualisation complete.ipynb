# HeatMap
#sns.heatmap() is used to create the heatmap.
iris.corr() computes the correlation matrix of the Iris dataset using the corr() function. This matrix contains the pairwise correlations between the numerical columns of the dataset.

annot=True displays the correlation values in each cell of the heatmap.

cmap='cubehelix' sets the color map of the heatmap to 'cubehelix', which is a colormap that varies smoothly between different colors.

linewidths=1 sets the width of the lines separating each cell in the heatmap to 1.

linecolor='k' sets the color of the lines separating each cell to black.

square=True makes each cell in the heatmap square-shaped.

mask=False does not mask any cells in the heatmap.

vmin=-1 and vmax=1 set the minimum and maximum values for the color scale, ensuring that the heatmap ranges from -1 to 1, representing the correlation values.

cbar_kws={"orientation": "vertical"} specifies the keyword arguments for the colorbar. In this case, it sets the orientation of the colorbar to vertical.

cbar=True displays a colorbar next to the heatmap.

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.heatmap(ir.corr(),annot=True,cmap='cubehelix',linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1,cbar_kws={"orientation": "vertical"},cbar=True)
```

```python
fig=plt.gcf()
fig.set_size_inches(12,8)
fig=sns.heatmap(ir.corr(),annot=True,cmap="cubehelix",linewidths=1,linecolor="k",square=True,mask=False,vmin=-1,vmax=1,cbar_kws={"orientation":"vertical"},cbar=True)
```
